```python
## replace missing values with mean for the selected column 

new_df2['latitude'] = new_df2['latitude'].fillna(new_df2['latitude'].mean())
new_df2['longitude'] = new_df2['longitude'].fillna(new_df2['longitude'].mean())
new_df2['depth'] = new_df2['depth'].fillna(new_df2['depth'].mean())
new_df2['mag'] = new_df2['mag'].fillna(new_df2['mag'].mean())
new_df2
```
